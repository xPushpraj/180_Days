#### Here we have the martix oprations

```python
a3 = np.array([
    [2,3,4],
    [3,4,5],
    [5,6,7]
])

a4 = np.arange(12).reshape(3,4)
print(a3)
print(a4)
```

```python
a3@a4                     # matrix product
```

```python
a3.dot(a4)                     # another matrix product
```
