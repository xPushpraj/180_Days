```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Create an array of x values (from 0 to 2*pi with 100 points)
x = np.linspace(0, 2 * np.pi, 100)

# 2. Compute the corresponding y values using the sine function
y = np.sin(x)

# 3. Plot the x and y values
plt.plot(x, y)

# 4. Add labels, a title, and a grid for clarity
plt.title("Sine Wave (y = sin(x))")
plt.xlabel("X values (radians)")
plt.ylabel("Amplitude")
plt.grid(True, which='both')
plt.axhline(y=0, color='k') # Add a horizontal line at y=0

# 5. Display the plot
plt.show()
```

##### Plot two different lines on the same graph with labels and legend.

```python
x = np.linspace(0,100,1000)
y1 = np.sin(x/10)*10
y2 = np.cos(x/10)*10

plt.plot(x, y1, color = 'green' , label = 'Sine Wave')
plt.plot(x, y2, color = 'red' , linestyle = '--' , label = 'Cos Wave')
```
