### expand_dims
##### Convert a 1D array into a column vector.

```python
a = np.random.randint(0,100,10)
a
```

```python
np.expand_dims(a,axis=1)
```

### Use of Where Function
##### Find indices where array elements are greater than a given value.

```python
np.where(a>30)    # it will result only the index numbers
```

```python
a[np.where(a>34)]   # it will result the values 
```

##### Replace values greater than a threshold using where.   (Condition , True , False)

```python
np.where(a>25, 'Yes', 'No')
```

```python
x = np.where(a>50, a, 0)
x
```

```python
arr = np.where(a>25, a, 0) | x
arr
```

##### Use where to filter rows from a 2D array.

```python
arr = np.random.randint(0,25,24).reshape(4,6)
arr
```

```python
z = np.where(arr>20 , 0 , arr)
z
```

### Argmax (argument of the maximum)

```python
a=np.argmax(arr)
a
```

```python
a=np.argmin(arr)
a
```

### Cumsum (cumulative sum)
